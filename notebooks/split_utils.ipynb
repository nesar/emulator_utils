### Read some spectra

```python
ell = readpolspice('../test/data/cl/power_247_30_mask_105_120.cl')[0]
fnames = glob('../test/data/cl/*')

cl_all = np.array([readpolspice(f)[1] for f in fnames])
```

### Divide the dataset into training and testing sets. 

```python
cl_train, cl_test, _, _ = random_holdout(cl_all, cl_all, split_fraction = 0.2)
```

```python
plt.figure(3)
plt.plot(ell, cl_train.T, '-', color='k', alpha=0.2, label='Train')
plt.plot(ell, cl_test.T, '-', color='r', alpha=0.4, label='Test')
plt.xscale('log')
plt.xlabel('l')
plt.ylabel('cl')
plt.legend()
```
